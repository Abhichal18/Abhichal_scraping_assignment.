```python
def get_webpage(url):
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    driver.get(url)
    time.sleep(5)
    webpage=driver.page_source
    return webpage
    
```

```python
def get_links(webpage,class_name,string=None):
    try:
        soup = BeautifulSoup(webpage, "html.parser")
        #print(soup)
        #print(soup.get_text())
        links = soup.find_all("a", string=string,attrs={'class':class_name})
        #print(links)
        links_list = []
        for link in links:
            item_link = link.get('href')
            #print(item_link)
            if URL not in item_link:
                links_list.append(URL+item_link)
            else:
                links_list.append(item_link)
    except:
        links_list=[]
    return links_list
```

```python
def get_product_count(webpage,class_name):
    soup = BeautifulSoup(webpage, "html.parser")
    links = soup.find("span", attrs={'class':class_name})
    return int(re.search(r'\d+',links.get_text()).group())
```

```python
def get_title(soup):
    try:
        title = soup.find('h1',attrs={'class': "ProductDetails_main__title__14Cnm"}).get_text()
    except:
        title = ""
    return title
```

```python
def get_price(soup):
    try:
        price =soup.find('span',attrs={'class': "ProductPrice_productPrice__price__3-50j"}).get_text().strip()
    except:
        price = ""
    return price
```

```python
def get_quantity(soup):
    try:
        quantity =soup.find('span',attrs={'class': "ProductPrice_productPrice__unit__2jvkA"}).get_text().strip()[1:]
    except:
        quantity = ""
    return quantity
```

```python
def get_characteristic(soup):
    try:
        quantity =soup.find('div',attrs={'class': "Tag_tag__1ynSO"}).get_text().strip()
    except:
        quantity = ""
    return quantity
```

```python
def get_description(soup):
    try:
        descript=soup.find('div', attrs={'class': "Expand_expand__container__3COzO"}).find_all('p')
        description = ' '.join(tag.get_text() for tag in descript)
    except:
        description=""
    return description
```

```python
def get_ingredients(soup):
    try:
        ingre=soup.find_all('li', attrs={'class': "IngredientsList_ingredientsList__item__1VrRy"})
        ingredients=[]
        for i in ingre:
            ingredients.append(i.get_text())
    except:
        ingredients=[]
    return ingredients
```

```python
def get_allergens(soup):
    try:
        allergens =soup.find('li',attrs={'class': "IngredientsSummary_ingredientsSummary__allergensListItem__2LBiz"}).get_text().strip()
    except:
        allergens = ""
    return allergens
```

```python
def get_interests(soup):
    try:
        inter=soup.find_all('span', attrs={'class': "FunTag_tag__text__1FfQ6"})
        interests=[]
        for i in inter:
            interests.append(i.get_text())
    except:
        interests=[]
    return interests
```

```python
def get_categories(soup):
    try:
        categ=soup.find_all('a', attrs={'class': "Link_link__1AZfr Breadcrumbs_list__link__rcn9k"})
        categories=[]
        for i in categ:
            categories.append(i.get_text())
    except:
        categories=[]
    return categories[2:]
```

```python
def get_image(soup,title):
    try:
        image = soup.find('img', attrs={'title':"Trader Joe's "+title}).get('src')
        image_url = URL+image
    except:
        image_url=""
    return image_url
```

```python
def get_serving_and_calory(soup):
    try:
        serve=soup.find_all('div', attrs={'class': "Item_characteristics__text__dcfEC"})
        serving_size=[]
        calories=[]
        for i in range(len(serve)):
            if i%2==0:
                serving_size.append(serve[i].get_text())
            else:
                calories.append(serve[i].get_text())
    except:
        serving_size,calories=[],[]
    return serving_size,calories
```

```python
d = {"product_id":[],"title":[],"description":[],"categories":[],"url":[], "price":[],
     "quantity":[],"image":[], "characteristics":[], "ingredients":[],"allergens":[],
     "interests":[],"servings":[],"calories":[]}
```

```python
for link in product_links:
    new_webpage=get_webpage(link)
    new_soup = BeautifulSoup(new_webpage, "html.parser")
    d['product_id'].append(link.split("/")[-1])
    d['title'].append(get_title(new_soup))
    d['description'].append(get_description(new_soup))
    d['categories'].append(get_categories(new_soup))
    d['url'].append(link)
    d['price'].append(get_price(new_soup))
    d['quantity'].append(get_quantity(new_soup))
    d['image'].append(get_image(new_soup,get_title(new_soup)))
    d['characteristics'].append(get_characteristic(new_soup))
    d['ingredients'].append(get_ingredients(new_soup))
    d['allergens'].append(get_allergens(new_soup))
    d['interests'].append(get_interests(new_soup))
    serving,calory=get_serving_and_calory(new_soup)
    d['servings'].append(serving)
    d['calories'].append(calory)
```

```python
traders_joy = pd.DataFrame.from_dict(d)
traders_joy = traders_joy.dropna(subset=['title'])
traders_joy.to_csv("output/traders_joy_data.csv", header=True, index=False)
```

## Validating Data

```python
df = pd.read_csv('output/traders_joy_data.csv',
                 converters={'categories':pd.eval,'ingredients':pd.eval,'interests':pd.eval,'servings':pd.eval,'calories':pd.eval})
```

```python
df=df.fillna("")
```

```python
data = df.to_dict(orient='records')
```

```python
len(data)
```

```python
validated_data=[]
invalid_data=[]
```

```python
def check(col_name, is_valid, invalid_columns):
    if not is_valid:
        invalid_columns.append(col_name)
    return invalid_columns
```

```python
for row in data:
    invalid_columns=[]
    is_valid_product_id = ProductValidator.validate_product_id(row['product_id'])
    invalid_columns=check('product_id',is_valid_product_id,invalid_columns)
    
    is_valid_title = ProductValidator.validate_title(row['title'])
    invalid_columns=check('title',is_valid_title,invalid_columns)
    #print(is_valid_title)
    
    is_valid_description = ProductValidator.validate_description(row['description'])
    invalid_columns=check('description',is_valid_description,invalid_columns)
    #print(is_valid_description)
    
    is_valid_url = ProductValidator.validate_url(row['url'])
    invalid_columns=check('url',is_valid_url,invalid_columns)
    #print(is_valid_url)
    
    is_valid_price = ProductValidator.validate_price(row['price'])
    invalid_columns=check('price',is_valid_price,invalid_columns)
    #print(is_valid_price)

    is_valid_quantity = ProductValidator.validate_quantity(row['quantity'])
    invalid_columns=check('quantity',is_valid_quantity,invalid_columns)
    #print(is_valid_price)

    is_valid_image_url = ProductValidator.validate_image_url(row['image'])
    invalid_columns=check('image',is_valid_image_url,invalid_columns)
    #print(is_valid_variants_images_urls)
    
    is_valid_ingredients = ProductValidator.validate_ingredients(row['ingredients'])
    invalid_columns=check('ingredients',is_valid_ingredients,invalid_columns)

    is_valid_servings = ProductValidator.validate_servings(row['servings'])
    invalid_columns=check('servings',is_valid_servings,invalid_columns)

    is_valid_calories = ProductValidator.validate_calories(row['calories'])
    invalid_columns=check('calories',is_valid_calories,invalid_columns)

    if len(invalid_columns)>0:
        row['invalid_columns']=invalid_columns
        invalid_data.append(row)
    else:
        validated_data.append(row)
```

```python
validated_data_df = pd.DataFrame.from_dict(validated_data)
```
